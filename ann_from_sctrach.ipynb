```python
def sigmoid(X):
    return 1/(1+np.exp(-X)) 
sigmoid(np.array([12,0,1]))
```

```python
def logloss(y_train,y_predicted):
    eplsion=1e-15
    y_predicted=[max(i,eplsion) for i in y_predicted]
    y_predicted=[min(i,1-eplsion)for i in y_predicted]
    y_predicted=np.array(y_predicted)

    return -np.mean(
        y_train*np.log(y_predicted)+(1-y_train)*np.log(1-y_predicted)
    )
    
```

```python
class mmy:
    def __init__(self):
        self.w1=1
        self.w2=1
        self.b=0
    def fit(self,x_train,y_train,ephos,threshold):
        self.w1,self.w2,self.b=self.gradient(x_train,y_train,ephos,threshold)
        
    def predict(self,x_varible):
        weight_sum=self.w1*x_varible.age+self.w2*x_varible.affordibility+self.b
        return sigmoid(weight_sum)
    def evaluate(self,x_test,y_test):
        y_predict=self.predict(x_test)
        error=np.mean((y_predict-y_test)**2)/2
        return error
        
    def gradient(self,x_train,y_train,ephos,threshold):
        w1=w2=1
        b=0
        rate=0.5
        n=len(x_train.age)
        for i in range(ephos):
            weight_sum=w1*x_train.age+w2*x_train.affordibility+b
            y_predicted=sigmoid(weight_sum)
            loss=logloss(y_train,y_predicted)
            w1d = (1/n)*np.dot(np.transpose(x_train.age),(y_predicted-y_train)) 
            w2d = (1/n)*np.dot(np.transpose(x_train.affordibility),(y_predicted-y_train)) 
    
            b1= np.mean(y_predicted-y_train)
            w1 = w1 - rate * w1d
            w2 = w2 - rate * w2d
            b=b-rate*b1
            print(f'w1={w1},w2={w2},b={b},loss={loss},ephos={i}')   
            if loss<=threshold:
                break
             
        return w1,w2,b
        
    
```

```python
instanet=mmy()
instanet.fit(x_train,y_train,5000,0.4231)
```
